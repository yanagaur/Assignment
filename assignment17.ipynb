```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.title("title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()

```

scatter plot is a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

With the subplot() function you can draw multiple plots in one figure

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
y1 = np.array([0, 1, 2, 3, 4, 5]) 
x1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([0, 1, 2, 3, 4, 5])
x2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([0, 1, 2, 3, 4, 5])
x3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([0, 1, 2, 3, 4, 5])
x4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot data in the first subplot
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('1')

# Plot data in the second subplot
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('2')

axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('3')

axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('4')


```

A bar plot, also known as a bar chart or bar graph, is a graphical representation of categorical data using rectangular bars. It is used to display and compare the values of different categories or groups. The length or height of each bar corresponds to the magnitude or frequency of the data being represented.

In a bar plot, the horizontal axis represents the categories or groups being compared, while the vertical axis represents the values or frequencies associated with each category. The bars are typically placed adjacent to each other, with equal width, and there is usually a gap between each bar to visually separate them.

Bar plots are particularly useful for showing comparisons between discrete categories or groups and are commonly used to illustrate and analyze data in various fields, such as statistics, economics, market research, and social sciences. They provide a clear visual representation of data and allow for easy interpretation and comparison of values across different categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create the box plot
plt.boxplot(data)

# Add labels and title
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot Example')

# Display the plot
plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of numerical data that provides a summary of its distribution. It displays key statistical measures, including the median, quartiles, and any potential outliers.
